# Computational Geometry

## 1. Orientation
Orientation of three ordered points $(P, Q, R)$ can be:
- **Counter-clockwise**
- **Clockwise**
- **Collinear**
It is determined by the sign of $(y_2 - y_1)(x_3 - x_2) - (y_3 - y_2)(x_2 - x_1)$.

## 2. Line Intersection
Two line segments $(P1, Q1)$ and $(P2, Q2)$ intersect if and only if:
- $(P1, Q1, P2)$ and $(P1, Q1, Q2)$ have different orientations.
- $(P2, Q2, P1)$ and $(P2, Q2, Q1)$ have different orientations.

## 3. Convex Hull
The smallest convex polygon that contains all given points.
### Algorithms:
- **Jarvis March**: $O(NH)$ where $H$ is the number of points on the hull.
- **Graham Scan**: $O(N \log N)$ using sorting by polar angle.
- **Monotone Chain**: $O(N \log N)$ by sorting by x-coordinates.
