# Persistent Segment Tree

A Persistent Segment Tree is a segment tree that allows us to access and query all of its past versions after updates. Instead of modifying existing nodes, it creates new nodes for the path that changed, sharing the rest of the nodes with the previous version.

### Key Concepts:
- **Immutability**: Every update creates a new root and a new path of $\log N$ nodes.
- **Space Efficiency**: Each update takes $O(\log N)$ extra space.
- **Querying History**: We can perform range queries on any version $V$ in $O(\log N)$ time.

### Applications:
- Finding the $k$-th smallest element in a range $[L, R]$.
- Solving geometric problems like counting points in a rectangle.
- Maintaining history of array states for undo/redo operations.

### Complexity:
- **Update**: $O(\log N)$ time and $O(\log N)$ space.
- **Query**: $O(\log N)$ time per version.
- **Initial Build**: $O(N)$ time and space.
