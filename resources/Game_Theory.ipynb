# Game Theory in Competitive Programming

## 1. Nim Game
The Nim Game is a classic game where two players take turns removing objects from heaps.
### Key Results:
- The position is a **losing position** (P-position) if and only if the XOR sum of all heap sizes is 0.
- Otherwise, it is a **winning position** (N-position).

## 2. Grundy Numbers (Sprague-Grundy Theorem)
Every impartial game under the normal play convention is equivalent to a Nim pile of a certain size. This size is the Grundy number (or nim-value) of the game state.

### Properties:
- $G(state) = mex(\{G(next\_state1), G(next\_state2), ...\})$
- **MEX (Minimum Excluded value)**: The smallest non-negative integer not present in the set.
- The XOR sum of Grundy numbers of independent subgames determines the overall game state's Grundy number.
